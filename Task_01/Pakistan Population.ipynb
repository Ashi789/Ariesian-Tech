```python
# Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# loading dataset
df = pd.read_csv('sub-division_population_of_pakistan.csv')
```

```python
df.head()
```

```python
print(df.columns)
```

```python
#Data Cleaning
df.isnull().sum()
```

```python
df = df.dropna(subset=['ALL SEXES (RURAL)', 'POPULATION 1998 (RURAL)', 'ALL SEXES (URBAN)', 'POPULATION 1998 (URBAN)'])
```

```python
df.describe()
```

```python
#Exploratory Data Analysis (EDA)
# Plot histogram for rural population distribution
plt.figure(figsize=(12, 6))
sns.histplot(df['ALL SEXES (RURAL)'], bins=30, kde=True)
plt.title('Population Distribution in Rural Areas')
plt.xlabel('Population')
plt.ylabel('Frequency')
plt.show()
```

```python
# Plot histogram for urban population distribution
plt.figure(figsize=(12, 6))
sns.histplot(df['ALL SEXES (URBAN)'], bins=30, kde=True)
plt.title('Population Distribution in Urban Areas')
plt.xlabel('Population')
plt.ylabel('Frequency')
plt.show()
```

```python
#For Rural Areas
current_year = 2024  # Use the actual current year if different

# Calculate annual growth rate for rural areas
df['ANNUAL_GROWTH_RATE (RURAL)'] = ((df['ALL SEXES (RURAL)'] - df['POPULATION 1998 (RURAL)']) / df['POPULATION 1998 (RURAL)']) * 100 / (current_year - 1998)
```

```python
#For Urban Areas
# Calculate annual growth rate for urban areas
df['ANNUAL_GROWTH_RATE (URBAN)'] = ((df['ALL SEXES (URBAN)'] - df['POPULATION 1998 (URBAN)']) / df['POPULATION 1998 (URBAN)']) * 100 / (current_year - 1998)
```

```python
#Verify 
print(df[['PROVINCE', 'ALL SEXES (RURAL)', 'POPULATION 1998 (RURAL)', 'ANNUAL GROWTH RATE (RURAL)', 'ALL SEXES (URBAN)', 'POPULATION 1998 (URBAN)', 'ANNUAL GROWTH RATE (URBAN)']].head())
```

```python
#Comparison of Growth Rates: Rural vs Urban
# Create a new column to indicate the type of area
df['AREA_TYPE'] = 'Rural'
df.loc[df['ANNUAL_GROWTH_RATE (URBAN)'].notna(), 'AREA_TYPE'] = 'Urban'

# Melt the DataFrame to make it suitable for seaborn
df_melted = df.melt(id_vars=['PROVINCE'], value_vars=['ANNUAL_GROWTH_RATE (RURAL)', 'ANNUAL_GROWTH_RATE (URBAN)'],
                    var_name='AREA_TYPE', value_name='ANNUAL_GROWTH_RATE')

# Update AREA_TYPE values
df_melted['AREA_TYPE'] = df_melted['AREA_TYPE'].replace({
    'ANNUAL_GROWTH_RATE (RURAL)': 'Rural',
    'ANNUAL_GROWTH_RATE (URBAN)': 'Urban'
})

plt.figure(figsize=(12, 6))
sns.boxplot(x='AREA_TYPE', y='ANNUAL_GROWTH_RATE', data=df_melted)
plt.title('Comparison of Growth Rates: Rural vs Urban')
plt.xlabel('Area Type (Rural/Urban)')
plt.ylabel('Annual Growth Rate (%)')
plt.show()
```

```python
#Annual Growth Rate by Province (Rural)
# Plot boxplot for annual growth rate by province (rural)
plt.figure(figsize=(12, 6))
sns.boxplot(x='PROVINCE', y='ANNUAL_GROWTH_RATE (RURAL)', data=df)
plt.title('Annual Growth Rate by Province (Rural)')
plt.xlabel('Province')
plt.ylabel('Annual Growth Rate (%)')
plt.xticks(rotation=90)
plt.show()
```

```python
#Annual Growth Rate by Province (Urban)
plt.figure(figsize=(12, 6))
sns.boxplot(x='PROVINCE', y='ANNUAL_GROWTH_RATE (URBAN)', data=df)
plt.title('Annual Growth Rate by Province (Urban)')
plt.xlabel('Province')
plt.ylabel('Annual Growth Rate (%)')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x='PROVINCE', y='ANNUAL_GROWTH_RATE (RURAL)', data=df)
plt.xticks(rotation=90)
plt.title('Annual Population Growth Rate by Province (Rural)')
plt.xlabel('Province')
plt.ylabel('Annual Growth Rate (%)')
plt.show()
```

```python
# Plotting annual growth rate by province for urban areas
plt.figure(figsize=(12, 6))
sns.barplot(x='PROVINCE', y='ANNUAL_GROWTH_RATE (URBAN)', data=df)
plt.xticks(rotation=90)
plt.title('Annual Population Growth Rate by Province (Urban)')
plt.xlabel('Province')
plt.ylabel('Annual Growth Rate (%)')
plt.show()
```

```python
# Check for missing values
print(df[['ANNUAL_GROWTH_RATE (RURAL)', 'ANNUAL_GROWTH_RATE (URBAN)', 'PROVINCE']].isna().sum())

# Drop rows with missing values if necessary
df = df.dropna(subset=['ANNUAL_GROWTH_RATE (RURAL)', 'ANNUAL_GROWTH_RATE (URBAN)'])
```

```python
# Calculate correlation matrix including growth rates for rural areas
corr = df[['ANNUAL_GROWTH_RATE (RURAL)', 'AREA (sq.km)', 'AVG HOUSEHOLD SIZE (RURAL)', 'SEX RATIO (RURAL)', 'SEX RATIO (URBAN)']].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation with Annual Growth Rate (Rural)')
plt.show()
```

```python
# Calculate correlation matrix including growth rates for urban areas
corr_urban = df[['ANNUAL_GROWTH_RATE (URBAN)', 'AREA (sq.km)', 'AVG HOUSEHOLD SIZE (URBAN)', 'SEX RATIO (URBAN)', 'SEX RATIO (RURAL)']].corr()
# Plot heatmap
sns.heatmap(corr_urban, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlation with Annual Growth Rate (Urban)')
plt.show()
```

```python
# Calculate correlation matrix including both urban and rural growth rates
corr_combined = df[['ANNUAL_GROWTH_RATE (RURAL)', 'ANNUAL_GROWTH_RATE (URBAN)', 
                    'AREA (sq.km)', 'AVG HOUSEHOLD SIZE (RURAL)', 'SEX RATIO (RURAL)', 
                    'AVG HOUSEHOLD SIZE (URBAN)', 'SEX RATIO (URBAN)']].corr()
# Plot heatmap
sns.heatmap(corr_combined, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlation Matrix: Urban vs. Rural Growth Rates')
plt.show()
```

```python

# Create a DataFrame with only Male, Female, and Transgender data
gender_df = df[['PROVINCE', 'MALE (RURAL)', 'MALE (URBAN)', 'FEMALE (RURAL)', 'FEMALE (URBAN)', 'TRANSGENDER (RURAL)', 'TRANSGENDER (URBAN)']]

# Melt the dataframe to make it easier for plotting
gender_df_melted = gender_df.melt(id_vars=['PROVINCE'], value_vars=['MALE (RURAL)', 'MALE (URBAN)', 'FEMALE (RURAL)', 'FEMALE (URBAN)', 'TRANSGENDER (RURAL)', 'TRANSGENDER (URBAN)'], var_name='Gender_Area', value_name='Population')

# Split Gender and Area into separate columns
gender_df_melted[['Gender', 'Area']] = gender_df_melted['Gender_Area'].str.split(' ', expand=True)
```

```python
# Plot Gender-wise comparison
plt.figure(figsize=(14, 6))
sns.barplot(x='PROVINCE', y='Population', hue='Gender_Area', data=gender_df_melted)
plt.title('Gender-wise Population Comparison: Urban vs Rural')
plt.xlabel('Province')
plt.ylabel('Population')
plt.xticks(rotation=90)
plt.legend(title='Gender & Area')
plt.show()
```

```python
plt.figure(figsize=(14, 6))
sns.boxplot(x='Gender_Area', y='Population', data=gender_df_melted)
plt.title('Gender-wise Population Distribution: Urban vs Rural')
plt.xlabel('Gender & Area')
plt.ylabel('Population')
plt.xticks(rotation=45)
plt.show()
```
