```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
data = pd.read_csv('Facebook Metrics of Cosmetic Brand.csv')

# Display the first few rows
data.head()
```

```python
# Check the basic info
data.info()
```

```python
# Summary statistics
data.describe()
```

```python

# Check for missing values
data.isnull().sum()
```

```python
# Set up the plotting area
fig, axes = plt.subplots(2, 2, figsize=(15, 10))

# Histogram of likes
sns.histplot(data['like'], kde=True, ax=axes[0, 0])
axes[0, 0].set_title('Distribution of Likes')

# Histogram of comments
sns.histplot(data['comment'], kde=True, ax=axes[0, 1])
axes[0, 1].set_title('Distribution of Comments')

# Histogram of shares
sns.histplot(data['share'], kde=True, ax=axes[1, 0])
axes[1, 0].set_title('Distribution of Shares')

# Histogram of Total Interactions
sns.histplot(data['Total Interactions'], kde=True, ax=axes[1, 1])
axes[1, 1].set_title('Distribution of Total Interactions')

plt.tight_layout()
plt.show()
```

```python
# Group by Type and calculate average engagement metrics
type_engagement = data.groupby('Type')[['like', 'comment', 'share', 'Total Interactions']].mean()

# Plot average engagement metrics by post type
type_engagement.plot(kind='bar', figsize=(12, 6))
plt.title('Average Engagement by Post Type')
plt.xlabel('Post Type')
plt.ylabel('Average Engagement')
plt.show()
```

```python
# Group by Post Month
month_engagement = data.groupby('Post Month')['Total Interactions'].mean()

# Plot engagement by month
plt.figure(figsize=(10, 6))
sns.lineplot(x=month_engagement.index, y=month_engagement.values, marker='o')
plt.title('Average Total Interactions by Month')
plt.xlabel('Month')
plt.ylabel('Average Total Interactions')
plt.show()


```

```python
# Group by Paid status and calculate average engagement metrics
paid_engagement = data.groupby('Paid')[['like', 'comment', 'share', 'Total Interactions']].mean()

# Plot average engagement metrics for paid vs. organic
paid_engagement.plot(kind='bar', figsize=(10, 6))
plt.title('Average Engagement for Paid vs. Organic Posts')
plt.xlabel('Paid (1 = Yes, 0 = No)')
plt.ylabel('Average Engagement')
plt.show()
```

```python
# Plot Total Reach vs Total Impressions
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Lifetime Post Total Reach', y='Lifetime Post Total Impressions', data=data)
plt.title('Total Reach vs Total Impressions')
plt.xlabel('Total Reach')
plt.ylabel('Total Impressions')
plt.show()

# Engagement Rate (Engaged Users / Total Reach)
data['Engagement Rate'] = data['Lifetime Engaged Users'] / data['Lifetime Post Total Reach']

```

```python
# Plot Engagement Rate distribution
plt.figure(figsize=(10, 6))
sns.histplot(data['Engagement Rate'], kde=True)
plt.title('Distribution of Engagement Rate')
plt.xlabel('Engagement Rate')
plt.show()
```

```python
# Group by Category and calculate average engagement
category_engagement = data.groupby('Category')['Total Interactions'].mean()

# Plot engagement by category
plt.figure(figsize=(10, 6))
category_engagement.plot(kind='bar')
plt.title('Average Total Interactions by Content Category')
plt.xlabel('Category')
plt.ylabel('Average Total Interactions')
plt.show()
```

```python
# Optional: Convert categorical columns to numeric using one-hot encoding
data_numeric = pd.get_dummies(data, columns=['Type', 'Category', 'Post Month', 'Post Weekday', 'Post Hour', 'Paid'], drop_first=True)
```

```python
# Select only numeric columns
numeric_data = data.select_dtypes(include=[np.number])
```

```python
# Correlation matrix
correlation_matrix = numeric_data.corr()

# Plot correlation matrix
plt.figure(figsize=(25, 15))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix')
plt.show()

```

```python
# Group by Post Hour and calculate average engagement
hourly_engagement = data.groupby('Post Hour')['Total Interactions'].mean()

# Plot average total interactions by hour
plt.figure(figsize=(12, 6))
sns.barplot(x=hourly_engagement.index, y=hourly_engagement.values, palette='viridis')
plt.title('Average Total Interactions by Post Hour')
plt.xlabel('Hour of the Day')
plt.ylabel('Average Total Interactions')
plt.show()
```

```python
# Scatter plot between Lifetime Post Total Reach and Lifetime Engaged Users
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Lifetime Post Total Reach', y='Lifetime Engaged Users', data=data)
plt.title('Lifetime Post Total Reach vs. Lifetime Engaged Users')
plt.xlabel('Lifetime Post Total Reach')
plt.ylabel('Lifetime Engaged Users')
plt.show()
```

```python
# Scatter plot between Lifetime Post Total Impressions and Total Interactions
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Lifetime Post Total Impressions', y='Total Interactions', data=data)
plt.title('Lifetime Post Total Impressions vs. Total Interactions')
plt.xlabel('Lifetime Post Total Impressions')
plt.ylabel('Total Interactions')
plt.show()
```

```python
# Group by Post Weekday and calculate average engagement
weekday_engagement = data.groupby('Post Weekday')['Total Interactions'].mean()

# Plot average total interactions by day of the week
plt.figure(figsize=(12, 6))
sns.barplot(x=weekday_engagement.index, y=weekday_engagement.values, palette='magma')
plt.title('Average Total Interactions by Day of the Week')
plt.xlabel('Day of the Week')
plt.ylabel('Average Total Interactions')
plt.xticks([0, 1, 2, 3, 4, 5, 6], ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun'])
plt.show()
```

```python
# Calculate Engagement Rate by Post Type
type_engagement_rate = data.groupby('Type')['Engagement Rate'].mean()

# Plot engagement rate by post type
plt.figure(figsize=(12, 6))
sns.barplot(x=type_engagement_rate.index, y=type_engagement_rate.values, palette='plasma')
plt.title('Average Engagement Rate by Post Type')
plt.xlabel('Post Type')
plt.ylabel('Average Engagement Rate')
plt.show()
```

```python
# Violin plot for distribution of likes by paid status
plt.figure(figsize=(12, 6))
sns.violinplot(x='Paid', y='like', data=data, palette='Set2')
plt.title('Distribution of Likes for Paid vs. Organic Posts')
plt.xlabel('Paid (1 = Yes, 0 = No)')
plt.ylabel('Number of Likes')
plt.show()
```

```python
# Violin plot for distribution of comments by paid status
plt.figure(figsize=(12, 6))
sns.violinplot(x='Paid', y='comment', data=data, palette='Set3')
plt.title('Distribution of Comments for Paid vs. Organic Posts')
plt.xlabel('Paid (1 = Yes, 0 = No)')
plt.ylabel('Number of Comments')
plt.show()
```

```python
# Seasonal trend: Total Interactions by Month
plt.figure(figsize=(12, 6))
sns.boxplot(x='Post Month', y='Total Interactions', data=data, palette='cool')
plt.title('Total Interactions Distribution by Month')
plt.xlabel('Month')
plt.ylabel('Total Interactions')
plt.show()
```

```python
# Seasonal trend: Total Interactions by Weekday
plt.figure(figsize=(12, 6))
sns.boxplot(x='Post Weekday', y='Total Interactions', data=data, palette='coolwarm')
plt.title('Total Interactions Distribution by Day of the Week')
plt.xlabel('Day of the Week')
plt.ylabel('Total Interactions')
plt.xticks([0, 1, 2, 3, 4, 5, 6], ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun'])
plt.show()
```

```python
# Scatter plot: Comments vs. Shares vs. Likes
plt.figure(figsize=(12, 6))
sns.scatterplot(x='comment', y='share', size='like', sizes=(20, 200), data=data, hue='Type', palette='deep', alpha=0.6)
plt.title('Comments vs. Shares vs. Likes by Post Type')
plt.xlabel('Number of Comments')
plt.ylabel('Number of Shares')
plt.legend(loc='upper right')
plt.show()
```

```python
# Scatter plot: Lifetime Post Consumers vs. Lifetime Engaged Users
plt.figure(figsize=(12, 6))
sns.scatterplot(x='Lifetime Post Consumers', y='Lifetime Engaged Users', data=data, hue='Type', palette='husl', alpha=0.7)
plt.title('Lifetime Post Consumers vs. Lifetime Engaged Users by Post Type')
plt.xlabel('Lifetime Post Consumers')
plt.ylabel('Lifetime Engaged Users')
plt.show()
```

```python
# Calculate engagement rate among page likers
data['Engagement Rate by Likers'] = data['Lifetime People who have liked your Page and engaged with your post'] / data['Lifetime Post reach by people who like your Page']

# Plot engagement rate by likers by post type
engagement_rate_by_likers = data.groupby('Type')['Engagement Rate by Likers'].mean()

plt.figure(figsize=(12, 6))
sns.barplot(x=engagement_rate_by_likers.index, y=engagement_rate_by_likers.values, palette='inferno')
plt.title('Average Engagement Rate Among Page Likers by Post Type')
plt.xlabel('Post Type')
plt.ylabel('Average Engagement Rate Among Page Likers')
plt.show()
```
